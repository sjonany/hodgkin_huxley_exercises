# 2.2 Restore the original conductance densities. Reduce the reversal potential (e.g. to 20mV) of Na to simulate a decrease in the extracellular Na concentration (i.e. external [Na+] is closer to internal [Na+]). What is the impact on the height/waveform of the action potential? Can a change in the conductance density of Na compensate for this (i.e. increase the height of the AP)? If not, why not?

Here is the original, where $E_{Na} = 50$

```python
model = HodgkinHuxley()
model.gen_plots(["V", "I", "gate"])
```

If we reduce $E_{Na}$ to 20, we see that the height of the action potential decreases.

```python
model = HodgkinHuxley()
model.E_Na = 20
model.gen_plots(["V", "I", "gate"])
```

```python
model = HodgkinHuxley()
model.E_Na = 20
model.g_Na *= 10
model.gen_plots(["V", "I", "gate"])
```

Increasing the conductance did help increase action potential height, but only up to a maximum voltage of 20. We will **not** be able to reach the initial max action potential of ~40 if we stick to the same current. This is because our bottleneck for V is $E_{Na}$. As soon as $V$ climbs up to $E_{Na}$, then all the ionic currents would have reversed in direction.

```python
model = HodgkinHuxley()
model.I_inj = lambda t: 900
model.E_Na = 20
model.g_Na *= 10
model.gen_plots(["V", "I", "gate"])
```

For funsies, increasing the injected current to an absurdly high amount would allow you to surpass $E_{Na} = 20$, but only for a short while before the gating effects catch up. This is because $\frac{dV}{dt} = I_{inj} + others$, so with a high enough $I_{inj}$, we should be able to make V increases as high as we want for a brief moment.
